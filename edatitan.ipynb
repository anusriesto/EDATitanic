now we are seeing a relation between age and passenger class so we ll try to relace the missing age value 
with relation with passenger class

```python
def impute_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
        
```

```python
mytrain["age"]=mytrain[["Age","Pclass"]].apply(impute_age,axis=1)   #one thing here i did not understand why[[]]
```
